```python
nfhs_dist.columns = nfhs_dist.columns.str.strip()
```

```python
suspicious_cols = []
for col in nfhs_dist.columns:
    if nfhs_dist[col].dtype == object:
        suspicious_cols.append(col)
```

```python
len(suspicious_cols)
```

```python
nfhs_dist[suspicious_cols[20:31]]
```

```python
for col in suspicious_cols[2:]:
    nfhs_dist[col] = nfhs_dist[col].replace("*", pd.NA)
```

```python
for col in suspicious_cols[2:]:
    nfhs_dist[col] = nfhs_dist[col].abs()
```

```python
for col in suspicious_cols[2:]:
    nfhs_dist[col] = pd.to_numeric(nfhs_dist[col], errors = "coerce")
```

```python
int_cols = []
for col in nfhs_dist.columns:
    if nfhs_dist[col].dtype == int:
        int_cols.append(col) 
```
