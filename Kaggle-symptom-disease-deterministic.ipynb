```python
all_symptoms = set()

for col in kaggle1.columns[1:]:
    vals = kaggle1[col]

    # cleaning each column
    vals = vals.dropna().astype(str).str.lower().str.strip()

    # removing empty entries
    vals = vals[vals != ""]

    # add to set
    all_symptoms.update(vals)
```

```python
len(all_symptoms)
```

```python
binary_df = pd.DataFrame(0, index = kaggle1.index, columns = sorted(all_symptoms))
```

```python
for i in kaggle1.index:
    symptoms = kaggle1.loc[i, kaggle1.columns[1:]].dropna()

    for s in symptoms:
        s = str(s).lower().strip()
        if s in binary_df.columns:
            binary_df.at[i, s] = 1
```

```python
binary_df["Disease"] = kaggle1["Disease"]
```
